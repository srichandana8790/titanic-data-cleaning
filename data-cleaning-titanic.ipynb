```python
titanic=pd.read_csv('/kaggle/input/datasets/shuofxz/titanic-machine-learning-from-disaster/train.csv')
titanic.head()
```

```python
titanic.info()
```

```python
titanic.isnull().head()
```

```python
titanic.isnull().sum()
```

```python
titanic['Age']=titanic['Age'].fillna(titanic['Age'].median())

titanic=titanic.drop(columns=['Cabin'])

titanic['Embarked']=titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])
```

```python
titanic.isnull().sum()
```

```python
print(titanic['Sex'].unique())

print(titanic['Embarked'].unique())
```

```python
print(titanic['Age'].describe())
print(titanic['Fare'].describe())
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = MinMaxScaler()
titanic['Fare_normalized'] = scaler.fit_transform(titanic[['Fare']])

scaler2 = StandardScaler()
titanic['Age_standardized'] = scaler2.fit_transform(titanic[['Age']])

titanic[['Age', 'Age_standardized', 'Fare', 'Fare_normalized']].head()
```

```python
print(titanic['Sex'].value_counts())

print(titanic['Embarked'].value_counts())
```

```python
titanic['Sex'] = titanic['Sex'].map({'male': 1, 'female': 0})

# Embarked - One Hot Encoding
titanic = pd.get_dummies(titanic, columns=['Embarked'])

# Result చూద్దాం
titanic[['Sex', 'Embarked_S', 'Embarked_C', 'Embarked_Q']].head()
```

```python
titanic['Embarked_S'] = titanic['Embarked_S'].astype(int)
titanic['Embarked_C'] = titanic['Embarked_C'].astype(int)
titanic['Embarked_Q'] = titanic['Embarked_Q'].astype(int)

titanic[['Sex', 'Embarked_S', 'Embarked_C', 'Embarked_Q']].head()
```

```python
titanic.head()
```

```python
titanic = titanic.drop(columns=['Age', 'Fare'])


print(titanic.shape)
titanic.head()
```

```python
titanic.to_csv('titanic_cleaned.csv', index=False)

print("File saved successfully!")
```
